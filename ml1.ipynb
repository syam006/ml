```python
#TASK-A
import pandas as pd 
df= pd.read_csv("labdata.csv")
print()
print(df)
print()
# first 5 elements
print(df.head())
#last 5 elements
print(df.tail())
#rows and columns 
rows,columns = df.shape
print(f"Rows: {rows} and cols : {columns}")
print(df.shape)
cols= df.columns
print(cols)
datatypes = df.dtypes
print(datatypes)
print(df.info())
```

```python
#TASK-B
att=df['Attendance'].mean()
print(f"mean of attendance: {att}")
print()
internalaverage = df['InternalMarks'].mean()
print(f"mean of internal marks: {internalaverage}")
print()
externalaverage = df['ExternalMarks'].mean()
print(f"mean of external marks: {externalaverage}")
print()
maxinter= df['InternalMarks'].max()
print(f"maximum internal marks: {maxinter}")
print()
mininter = df['InternalMarks'].min()
print(f"minimum internal marks: {mininter}")
print()

maxexternal= df['ExternalMarks'].max()
print(f"maximum external marks: {maxexternal}")
print()

minexternal = df['ExternalMarks'].min()
print(f"minimum external marks: {minexternal}")
print()

stdofinternal = df['InternalMarks'].std()
print(f"standard deviation of internal marks: {stdofinternal}")
print()

print(df.describe())
```

```python
#TASK-C

totalmarks = df['InternalMarks']+ df['ExternalMarks']
df['TotalMarks'] = totalmarks

percentage = (df['TotalMarks']/100)*100
df['Percentage']= percentage
print(df)

```

```python
#TASK-D

attgreat90=df[df['Attendance']>90]
print("Students with attendance greater than 90%:")
print(attgreat90)
print()

totexc90 = df[df['TotalMarks']>90]
print("Students with total marks greater than 90:")
print(totexc90)
print()

femalestudents = df[df['Gender']=='F']
print("Female students:")
print(femalestudents)
print()

malestudents = df[df['Gender']=='M']
print("Male students:")
print(malestudents)
print()

exterless50 = df[df['ExternalMarks']<50]
print("Students with external marks less than 50:")
print(exterless50)
print()
```

```python
#TASK-E

stuhighest = df[df['TotalMarks']==df['TotalMarks'].max()]
print("Student(s) with the highest total marks:")
print(stuhighest)
print()

stulowest = df[df['TotalMarks']==df['TotalMarks'].min()]
print("Student(s) with the lowest total marks:")
print(stulowest)


top3 = df.nlargest(3, 'TotalMarks')
print("Top 3 students with the highest total marks:")
print(top3)
print()

last3 = df.nsmallest(3, 'TotalMarks')
print("Last 3 students with the lowest total marks:")   
print(last3)
print()

aboveclassaverage = df[df['TotalMarks']>df['TotalMarks'].mean()]
print("Students with total marks above the class average:")
print(aboveclassaverage)
print()
```

```python
#task-f

averagemale = df[df['Gender']=='M']['TotalMarks'].mean()
print(f"average marks for male students: {averagemale}")
print()


averagefemale = df[df['Gender']=='F']['TotalMarks'].mean()
print(f"average marks for female students: {averagefemale}")
print()

compareattmf = df.groupby('Gender')['Attendance'].mean()
print("Average attendance for male and female students:")
print(compareattmf)
```

```python
#task-e

import matplotlib.pyplot as plt

barcharttotalmarks = df.plot.bar(x='Name', y='TotalMarks', title='Total Marks of Students', legend=False)
plt.show()


histogram = df['Attendance'].plot.hist(title='Attendance Histogram', bins=10)
plt.xlabel('Attendance')

piechartgender = df['Gender'].value_counts().plot.pie(title='Gender Distribution', autopct='%1.1f%%')
plt.ylabel('')

scatterplot = df.plot.scatter(x='InternalMarks', y='ExternalMarks', title='Internal vs External Marks')
plt.show()

```
